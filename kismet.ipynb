```python
def kismet(util_mat):
    K = util_mat.copy()
    n = util_mat.shape[0]
    m = util_mat.shape[1]
    for i in range(n):
        for j in range(m):
            if K[i,j] > 3:
                K[i,j] = 1
            else:
                K[i,j] = 0

kismet(utility_matrix_zeros)
```
